```python
"This is where we will combine datasets and clean them"
import pandas as pd
```

```python
records = pd.read_csv('./college_football_records.csv')
scorecard = pd.read_csv('./combined_scorecard_data.csv')
```

```python
records['year'] = records['year'] + 2
```

```python
#check how different university names are
records_univs = set(records['team'].unique())
scorecard_univs = set(scorecard['INSTNM'].unique())
```

```python
from thefuzz import process

name_mapping = {}
for name in records_univs:
    best_match, score = process.extractOne(name, scorecard_univs)
    if score > 80 and best_match in scorecard_univs:
        name_mapping[name] = best_match

# Apply the mapping
records['team'] = records['team'].replace(name_mapping)

# Print any unmatched universities for manual review
unmatched_universities = records_univs - set(name_mapping.keys())
print("Unmatched Universities:", unmatched_universities)
```

```python
merged_df = scorecard.merge(records, how='inner', left_on=['INSTNM', 'Year'], right_on=['team', 'year'])
```

```python
merged_df.drop(['team', 'year', 'scorecard_name'], axis=1, inplace=True)
```

```python
merged_df['win_percentage'] = merged_df['won']/(merged_df['won'] + merged_df['lost'])
```
